```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import Imputer
```

```python
with open("C:\\Users\\Sujina\\Downloads\\nba_2013.csv", 'r') as csvfile:
    nba = pd.read_csv(csvfile)
```

```python
nba['bref_team_id'],_=pd.factorize(nba['bref_team_id'])
```

```python
nba['pos'],_=pd.factorize(nba['pos'])
```

```python
X=nba.iloc[:,1:28]
y=nba.iloc[:,28]
```

```python
imp = Imputer(missing_values="NaN", strategy='median', axis=0)
X = imp.fit_transform(X)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsRegressor
from sklearn.metrics import r2_score
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, random_state = 0)
#y_train = y_train.ravel()
#y_test = y_test.ravel()
```

```python
for K in range(1,25):
    
    neigh =KNeighborsRegressor(n_neighbors = K, weights='uniform', algorithm='auto')
    neigh.fit(X_train, y_train)
    y_pred = neigh.predict(X_test)
    acc=r2_score(y_test,y_pred)
    print("Accuracy is ", acc*100,"% for K-Value:",K)
```
