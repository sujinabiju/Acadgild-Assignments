```python
import numpy as np
import pandas as pd
import sqlite3
from sqlalchemy import create_engine
engine1 = create_engine('sqlite:///:memory:', echo=False)
```

```python
data=pd.read_csv("https://archive.ics.uci.edu/ml/machine-learning-databases/adult/adult.data", header=None)
data.head()
```

```python
data.columns=['age','workclass','fnlwgt','education','education-num','marital-status',
                     'occupation','relationship','race','sex','capital-gain','capital-loss',
                     'hours-per-week','native-country','class']
data.head(10)
```

```python
data.to_sql('sqladb', con=engine1, if_exists='replace',chunksize=60)
```

```python
engine1.execute("SELECT * FROM sqladb LIMIT 10;").fetchall()
```

```python
engine1.execute("select avg([hours-per-week]) from sqladb where sex == ' Male' and workclass == ' Private'").fetchall()
```

```python
engine1.execute("SELECT AVG([hours-per-week]) as Average FROM sqladb WHERE Workclass=='Private' ").fetchall()
```

```python
engine1.execute("select distinct(education),count(education) from sqladb group by education;").fetchall()
```

```python
engine1.execute("select occupation,count(occupation) from sqladb group by occupation").fetchall()
```

```python
engine1.execute("select relationship,count(relationship) from sqladb group by relationship").fetchall()
```

```python
q="""
select * from sqladb 
where [marital-status] != ' Never-married' 
and workclass == ' Private' 
and education == ' Masters'
"""
engine1.execute(q).fetchall()
```

```python
q="""
select workclass, avg(age),min(age),max(age)
from sqladb
group by workclass
"""
engine1.execute(q).fetchall()
```

```python
engine1.execute("select age,[native-country],count(age) from sqladb group by age,[native-country] ").fetchall()
```

```python
q="""
alter table sqladb
add column 'Net-Capital-Gain' BIGINT
"""
q1="""
UPDATE sqladb
SET 'Net-Capital-Gain' = 'capital-gain' - 'capital-loss' 
"""
engine1.execute(q)
engine1.execute(q1)
engine1.execute("select * from sqladb limit 5").fetchall()
```

```python
import numpy as np
import pandas as pd
import sqlite3
from sqlalchemy import create_engine
engine = create_engine('sqlite://', echo=False)
```

```python
data1=pd.read_csv("https://archive.ics.uci.edu/ml/machine-learning-databases/adult/adult.data", header=None)
data1.head()
```

```python
data1.columns=['age','workclass','fnlwgt','education','education-num','marital-status',
                     'occupation','relationship','race','sex','capital-gain','capital-loss',
                     'hours-per-week','native-country','class']
data1.head()
```

```python
data1.to_sql('adult', con=engine,if_exists='replace',chunksize=60)
```

```python
engine.execute("UPDATE adult SET [capital-gain] = 2000 ;")
```

```python
engine.execute("UPDATE adult SET fnlwgt = 70000 where age > 50 ;")
```

```python
engine.execute("select * from adult limit 10").fetchall()
```

```python
engine.execute("delete from adult where age = 90")
```

```python
engine.execute("delete from adult where education in (' 11th',' 12th')")
```

```python
engine.execute("select education,count(education) from adult group by education;").fetchall()
```

```python
engine.execute("select * from adult where education=' Masters' and age <30 and sex = ' Male'").fetchall()
```

```python
engine.execute("select max([capital-gain]) from  adult").fetchall()
```

```python
q="""
select count([native-country]) from adult where [native-country]=' India' group by [native-country]
"""
engine.execute(q).fetchall()
```
